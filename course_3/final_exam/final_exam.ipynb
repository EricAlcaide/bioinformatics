```python
# Answer to Question#3 - ORFs features?
def read_seq(seq, frame=3, printer=True):
    ORF = []
    aux = 0
    seq = [seq[i:i+3] for i in range(frame-1, len(seq), 3)]
    dna = seq[:]
    n_starts = seq.count("ATG")
    if printer: print("\nPotential number of ORFs: ", n_starts, "\n")
    # Find all ORFs
    while "ATG" in seq:
        # Find starting position
        start = seq.index("ATG")
        seq = seq[start+1:]
        aux += start
        # End the loop if no stop codon is found
        try: stop = min([seq.index(stop) for stop in ["TAA", "TAG", "TGA"] if stop in seq])
        except: break
        # Record ORFs
        pre_orf = seq[:stop+1]
        pre_orf.insert(0, "ATG")
        ORF.append([aux*3+frame-1, ''.join(pre_orf)])
        if printer: print([aux*3+frame-1, aux*3+frame-1+stop*3, ''.join(pre_orf)])
        aux += 1
        
    if printer: print("\nTotal number of ORFs is:", len(ORF))
    return ORF

def find_orf_file(seqs, frame=3):
    orfs = {}
    total_orfs = 0
    for k in seqs.keys():
        seq_orf = read_seq(seqs[k], frame=frame, printer=False)
        orfs[k] = seq_orf
        print(k, len(seq_orf))
        total_orfs += len(seq_orf)
    
    return orfs, total_orfs

def longest_orf(seq, frame=None):
    if not frame: orfs = [read_seq(seq, frame=n+1, printer=False) for n in range(3)]
    else: orfs =  [read_seq(seq, frame=frame, printer=False)]
    longest = {"length": 0, "dna": "", "start":0}
    for group in orfs:
        for orf in group:
            if len(orf[1])>longest["length"]:
                longest["length"] = len(orf[1])
                longest["dna"] = orf[1]
                longest["start"] = orf[0]+1
    return longest

def seq_longest_orf(seqs, frame=None):
    longest = {"seq_id": "", "orf": {"length": 0, "dna": "", "start":0}}
    for k in seqs.keys():
        partial = longest_orf(seqs[k], frame=frame)
        if partial["length"] > longest["orf"]["length"]:
            longest["seq_id"] = k
            longest["orf"] = partial
    return longest
```

```python
# Answer to Question#4 - Repeats in forward strands?
def find_repeat_seq(seq, l):
    seen, counter = set(), {}
    for i in range(len(seq)-l):
        part = seq[i:i+l]
        if part in seen:
            counter[part] += 1
        else: 
            seen.add(part)
            counter[part] = 1
    
    return counter  

def find_repeat_multiseq(seqs, l):
    seen, counter = set(), {}
    for seq in seqs.values():
        for i in range(len(seq)-l):
            part = seq[i:i+l]
            if part in seen:
                counter[part] += 1
            else: 
                seen.add(part)
                counter[part] = 1
    
    return counter  
```
